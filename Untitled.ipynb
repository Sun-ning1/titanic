```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2) 
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)#training the logistic regresstion model with train_df
```

```python
X_train_prediction = model.predict(X_train)#result
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)#compare
print('accuracy score: ',training_data_accuracy)
```
